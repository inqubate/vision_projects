```python
import numpy as np
import tensorflow as tf
import random as python_random
import os
os.environ["PYTHONHASHSEED"]="0"
np.random.seed(123)
python_random.seed(123)
tf.random.set_random_seed(123)
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data(path="mnist.npz")

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

#print("Y", y_train[0])

x_train = x_train / 255
x_test = x_test / 255

#print("X", x_train[0])
```

<h3>Task 1</h3>

Hypothesis:

Description: Create the starting point of part 1, which different models can be compared to.
    
Input:
Number of nodes for the hidden layer: Changed from 8 to 32
Output change: The model's accuracy as well as the training time should increase.
Change reason: More parameters should allow the model to capture more detail but more parameters means more values to train so it should take longer.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD

model = Sequential()
model.add(Flatten()) 
model.add(Dense(32, activation='sigmoid')) 
model.add(Dense(10, activation='softmax'))
model.compile(loss="categorical_crossentropy", optimizer=SGD(), metrics=['accuracy'])
```

```python
epochs = 5
mini_batch_size = 128
for epoch in range(epochs):
    model.fit(x_train, y_train, batch_size=mini_batch_size)
```

```python
score, acc = model.evaluate(x_test, y_test)
print('Test score:', score)
print('Test accuracy:', acc)
```

Output:

Results: Training time was about 5 seconds.

Changes in results: Test accuracy improved to 81.8 % but also the training time was quicker than in the prework.
Reasoning: No idea why training was quicker with a larger model.

<h3>Task 2</h3>

```python
#create_dense() - a function that takes an array of sizes for the hidden layers as an argument. 
#Creates a multi-layer perceptron that always has appropriate input and output layers for the MNIST task
def create_dense(sizes):
    model = Sequential()
    model.add(Flatten(input_shape=(28,28)))
    for s in sizes:
        model.add(Dense(s, activation='sigmoid')) 
    model.add(Dense(10, activation='softmax'))
    model.compile(loss="categorical_crossentropy", optimizer=SGD(), metrics=['accuracy'])
    return model
```

```python
import matplotlib.pyplot as plt

#evaluate() - a function that prints a summary of the model, 
#trains the model, graphs the training and validation accuracy, 
#and prints a summary of its performance on the test data
def evaluate(model, epochs=5, mini_batch_size=128):
    print(model.summary())
    #for epoch in range(epochs):
    history = model.fit(x_train, y_train, batch_size=mini_batch_size, epochs=epochs)
    #print(history.history.keys())
    plt.plot(history.history['acc'])
    #plt.plot(history.history['val_accuracy']) #Not using a validation set - I don't think MNIST comes with one by default(?) - if turned on validation, the dev set would change and results would not be consistent.
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()
    score, acc = model.evaluate(x_test, y_test)
    print('Test score:', score)
    print('Test accuracy:', acc)
```

<b>Case 1</b>

Hypothesis:

Description: Deep networks (e.g., different number of hidden layers while keeping nodes constant); use 32 nodes so results are comparable with Task 1 results.
    
Input:
Number of hidden layers: Changed from 1 to 2. Size of each is 32.
Output change: The model's accuracy as well as the training time should increase.
Change reason: More parameters should allow the model to capture more detail but more parameters means more values to train so it should take longer.

```python
model = create_dense([32, 32])
evaluate(model)
```

Output:

Results: Accuracy decreased significantly to only 50.0 %. Training time increased only very marginally.

Changes in results: Maybe the model with one more hidden layer would require more training to achieve a more respectable level of accuracy. Larger models require more training as they have more parameters that need adjusting.
